## 1. Fibonacci Number




```python
import math

def fibonacciNumber (n):
    factor = ((1 + math.sqrt(5))/2)
    return int((1/math.sqrt(5)) * (factor**n - (1-factor)**n))

n = int(input())
print(fibonacciNumber(n))
```

## 2. Last Digit of a Large Fibonacci Number

```python
import math

def lastDigitFibonacci(n):
    prev = 0
    actual = 1

    if n <= 1:
        return n 
    for i in range (n - 1):
        next = prev + actual
        prev = actual
        actual = next
    
    return next % 10

n=int(input())
print(lastDigitFibonacci(n))
```

## 5.  Fibonacci Number Again

```python
import sys
def pisano(m):
    prev = 0
    act = 1
    next = prev + act

    for i in range(m * m):
        next = (prev + act) % m
        prev = act
        act = next
        if (prev == 0 and act == 1):
            return i + 1


def fibonacci(n, m):
    var_pisano = n % pisano(m)
    prev = 0
    act = 1
    newPisano = var_pisano

    for i in range(var_pisano - 1):
        newPisano = (prev + act) % m
        prev = act
        act = newPisano

    result_module = newPisano % m
    return result_module

input = sys.stdin.read()
# a, b = map(int, input.split())
a = 239
b = 1000
print(fibonacci(a,b))
```

## 6. Last Digit of the Sum of Fibonacci Numbers

```python
def lastDigitOfSumFibonacci(n):
    prev = 0
    actual = 1
    
    if (n == 0):
        return 0

    if (n == 1):
        return 1

    else:
 
        # Since pi(10)=60, the last digit of F_n repeats with period 60
        rem = n % 60
 
        if(rem == 0):
            return 0
 
        for i in range(2, rem + 3):
            next =(prev + actual) % 60
            prev = actual
            actual = next
 
        sum = actual - 1
        return (sum % 10)

# n = int(input())
n = 100
print(lastDigitOfSumFibonacci(n))
```

## 7.  Last Digit of the Sum of Fibonacci Numbers Again

```python
import sys

def fibonacci(n):
    prev = 0
    actual = 1
    for i in range(2, n+1):
        next = (prev + actual) % 10
        prev = actual
        actual = next

    return next - 1

input = sys.stdin.read()
#n, m = map(int, input.split())
n = 10
m = 200
# Since pi(10)=60, the last digit of F_n repeats with period 60
ln = (n + 2) % 60
lm = (m + 1) % 60

if ln<=1:
    a = ln - 1
else:
    a = fibonacci(ln)

if lm<=1:
    b = lm-1
else:
    b = fibonacci(lm)


if a>=b:
    print(a-b)
else:
    print(10 + a-b)
```
